# 2.9 white noise

Time series that show no autocorrelation are called white noise. Figure 2.17 gives an example of a white noise series.


```r
set.seed(30)
y <- ts(rnorm(50))
autoplot(y) + ggtitle("White noise")
```

```r
ggAcf(y)
```

For a white noise series, we expect 95% of the spikes in the ACF to lie within  
±
2
/
√
T
  where  
T
  is the length of the time series. 

If one or more large spikes are outside these bounds, or if substantially more than 5% of spikes are outside these bounds, then the series is probably not white noise.
